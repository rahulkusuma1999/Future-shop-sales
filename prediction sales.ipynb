```python
#data leakage technique
#when the data you are using to train a machine learning algorithm happens to have the information you are trying to predict
#if any other feature whose value would not actually be available in practice at the time you’d want to use the model to make a prediction, is a feature that can introduce leakage to your model
test_shop_ids = test['shop_id'].unique()
test_item_ids = test['item_id'].unique()
# Only shops that exist in test set.
leak_train = sales_train[sales_train['shop_id'].isin(test_shop_ids)]
# Only items that exist in test set.
leak_train = leak_train[leak_train['item_id'].isin(test_item_ids)]
```

```python
shop_ids = train['shop_id'].unique()
item_ids = train['item_id'].unique()
empty_df = []
for i in range(34):
    for shop in shop_ids:
        for item in item_ids:
            empty_df.append([i, shop, item])

empty_df = pd.DataFrame(empty_df, columns=['date_block_num','shop_id','item_id'])
```

```python
train = pd.merge(empty_df, train, on=['date_block_num','shop_id','item_id'], how='left')
train.fillna(0, inplace=True)
```

```python
train_set = train.query('date_block_num >= 0 and date_block_num < 26').copy()
validation_set = train.query('date_block_num >= 26 and date_block_num < 33').copy()
test_set = train.query('date_block_num == 33').copy()
```

```python
# Creating training and validation sets
x_train = train_set.drop(['mitem_cnt','date_block_num'],axis=1)
y_train = train_set['mitem_cnt'].astype(int)

x_val = validation_set.drop(['mitem_cnt','date_block_num'],axis=1)
y_val = validation_set['mitem_cnt'].astype(int)
```

```python
latest_records = pd.concat([train_set, validation_set]).drop_duplicates(subset=['shop_id', 'item_id'], keep='last')
x_test =pd.merge(test, latest_records, on=['shop_id', 'item_id'], how='left', suffixes=['', '_'])
x_test['year'] = 2015
x_test['month'] = 9
x_test.drop('mitem_cnt', axis=1, inplace=True)
x_test = x_test[x_train.columns]
```

```python
sets = [x_train, x_val, x_test]
for dataset in sets:
    for shop_id in dataset['shop_id'].unique():
        for column in dataset.columns:
            shop_median = dataset[(dataset['shop_id'] == shop_id)][column].median()
            dataset.loc[(dataset[column].isnull()) & (dataset['shop_id'] == shop_id), column] = shop_median
            
# Fill remaining missing values on test set with mean.
x_test.fillna(x_test.mean(), inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from scipy import stats
from sklearn.linear_model import ridge_regression
from sklearn.svm import LinearSVC
```

```python
# These will be our base models
m1 = LinearRegression()
m2 = DecisionTreeRegressor()
m3 = RandomForestRegressor(n_estimators=10)
m4= LinearSVC()
```
